```python
#Bar Chart
plt.figure(figsize=(10, 5))
sns.barplot(x=df["Item"].value_counts().index, y=df["Item"].value_counts().values, palette="viridis")
plt.title(" Number of Orders per Product")
plt.xlabel("Product")
plt.ylabel("Number of Orders")
plt.show()
```

```python
#Pie Chart 
plt.figure(figsize=(6, 6))
plt.pie(df["Payment Method"].value_counts(), labels=df["Payment Method"].value_counts().index, 
        autopct='%1.1f%%', colors=["red", "blue", "green"])
plt.title(" Payment Method Distribution", fontsize=15, fontweight='bold')
plt.legend()
plt.show()
```
