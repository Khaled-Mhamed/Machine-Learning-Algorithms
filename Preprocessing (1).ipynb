```python
import pandas as pd
```

```python
df = pd.read_csv('dirty_cafe_sales.csv')
```

```python
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors='coerce')  # يحول "ERROR" إلى NaN
```

```python
df.replace("UNKNOWN", df.mode().iloc[0], inplace=True)
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors='coerce')
```

```python
df.dropna(inplace=True)
```

```python
df = df[df["Payment Method"] != "ERROR"]
```

```python
df.isnull().sum()
```

```python
cleaned_file_path = "cleaned_cafe_sales.csv"
```

```python
df.to_csv(cleaned_file_path, index=False)
```
